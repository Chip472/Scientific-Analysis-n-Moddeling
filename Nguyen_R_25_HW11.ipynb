1. Write a ``Particle`` class that can be used to represent a particle with a mass, a 3-d position, and a 3-d velocity.

2. Write a method that can be used to compute the kinetic energy of the particle

3. Write a method that takes another particle as an argument and finds the distance between the two particles

4. Write a method that given a time interval ``dt`` will update the position of the particle to the new position based on the current position and velocity.

5. Write a ``ChargedParticle`` class that inherits from the ``Particle`` class, but also has an attribute for the charge of the particle.


```python
import numpy as np
# Copy your Particle class here
class Particle(object):
    def __init__(self, mass, x, y, z, vx, vy, vz):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz

    def kineticE(self):
        return 0.5 * self.mass * (self.vx**2 + self.vy**2 + self.vz**2)

    def distance(self, other):
        return np.sqrt((self.x - other.x)**2 + (self.y - other.y)**2 + (self.z - other.z)**2)

    def newPosition(self, dt):
        self.x = self.x + self.vx * dt
        self.y = self.y + self.vy * dt
        self.z = self.z + self.vz * dt
        return (self.x, self.y, self.z)


# part 5
class ChargedParticle(Particle):
    def __init__(self, mass, x, y, z, vx, vy, vz, charge):
        super().__init__(mass, x, y, z, vx, vy, vz)
        self.charge = charge

    def chargeP(self):
        return self.charge
```

```python
# write/copy code here

particle1 = Particle(0.5,1,2,3,6,5,4)
print(particle1.kineticE())

particle2 = Particle(0.3,2,6,7,9,2,5)
print(particle1.distance(particle2))

print(particle1.newPosition(0.1))

particle3 = ChargedParticle(4,3,5,1,3,5,2,7)
print(particle3.chargeP())
```

```python
# write new code here

class ChargedParticle(Particle):
    def __init__(self, mass, x, y, z, vx, vy, vz, charge):
        super().__init__(mass, x, y, z, vx, vy, vz)
        self.charge = charge

    def chargeP(self):
        return self.charge

    def interaction(self, other):
        if self.distance(other) < 0.25:
            print("interaction")
            if self.charge * other.charge < 0:
                print("merge")
                self.vx = self.vy = self.vz = 0
                other.vx = other.vy = other.vz = 0
                self.charge = 0
                other.charge = 0
            else:
                print("repel")
                factor_self = -1 * (self.charge+other.charge) * (self.mass/other.mass)
                factor_other = -1 * (self.charge+other.charge) * (other.mass/self.mass)

                self.vx *= factor_self
                self.vy *= factor_self
                self.vz *= factor_self

                other.vx *= factor_other
                other.vy *= factor_other
                other.vz *= factor_other


p1 = ChargedParticle(1,-5,-5,-5,1,1,1,0.5)
p2 = ChargedParticle(1,5,5,5,-1,-1,-1,-0.5)

dt = 0.05

for step in range(300):
    p1.newPosition(dt)
    p2.newPosition(dt)
    p1.interaction(p2)

    print("Step:", step)
    print("P1 position:", p1.x, p1.y, p1.z, "velocity:", p1.vx, p1.vy, p1.vz, "charge:", p1.charge)
    print("P2 position:", p2.x, p2.y, p2.z, "velocity:", p2.vx, p2.vy, p2.vz, "charge:", p2.charge)
    print(" ")

p1 = ChargedParticle(1,-5,-5,-5,2,2,2,0.5)
p2 = ChargedParticle(1,5,5,5,-1,-1,-1,2)

for step in range(300):
    p1.newPosition(dt)
    p2.newPosition(dt)
    p1.interaction(p2)

    print("Step:", step)
    print("P1 position:", p1.x, p1.y, p1.z, "velocity:", p1.vx, p1.vy, p1.vz, "charge:", p1.charge)
    print("P2 position:", p2.x, p2.y, p2.z, "velocity:", p2.vx, p2.vy, p2.vz, "charge:", p2.charge)
    print(" ")
```
